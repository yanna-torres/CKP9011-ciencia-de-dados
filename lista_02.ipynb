```python
import pandas as pd
import numpy as np
import duckdb
```

## a) Ler o dataset fakeTelegram.BR_2022.csv

```python
# Mesmo link utilizado na lista 01 para acessar os dados
data_url = "https://media.githubusercontent.com/media/yanna-torres/CKP9011-ciencia-de-dados/refs/heads/lista-01/data/fakeTelegram.BR_2022.csv"

df = pd.read_csv(data_url)
```

## b) Remova os trava-zaps

## c) Exportar os dados para um arquivo Parquet

```python
df.to_parquet("fakeTelegram.BR_2022.parquet")
```

## d) Exportar os dados para o DuckDB

```python
conn = duckdb.connect()

conn.execute("CREATE TABLE messages AS SELECT * FROM df")
```

```python
result = conn.execute("SELECT * FROM messages").fetchone()
```

```python
result
```

## e) Utlizando o DuckDB recupere

### 1. A quantidade de mensagens

```python
quant_messages = conn.execute("SELECT COUNT(*) AS total_messages FROM messages").fetchone()
```

```python
quant_messages
```

### 2. A quantidade de usuários

```python
quant_users = conn.execute("SELECT COUNT(DISTINCT id_member_anonymous) AS total_users FROM messages").fetchone()
```

```python
quant_users
```

### 3. A quantidade de grupos

```python
quant_groups = conn.execute("SELECT COUNT(DISTINCT id_group_anonymous) AS total_groups FROM messages").fetchone()
```

```python
quant_groups
```

### 4. Quantidade de mensagens que possuem apenas texto

```python
quant_messages_only_text = conn.execute("SELECT COUNT(*) AS total_messages FROM messages WHERE text_content_anonymous IS NOT NULL AND has_media = False").fetchone()
```

```python
quant_messages_only_text
```

### 5. Quantidade de mensagens contendo mídias

```python
quant_messages_only_media = conn.execute("SELECT COUNT(*) AS total_messages FROM messages WHERE text_content_anonymous IS NULL AND has_media = True").fetchone()
```

```python
quant_messages_only_media
```

### 6. Quantidade de mensagens por tipo de mídia (jpg, mp4 etc)

```python
quant_messages_per_media_type_df = conn.execute("""
    SELECT media_type, COUNT(*) AS total_messages
    FROM messages
    WHERE has_media = True OR media IS NOT NULL
    GROUP BY media_type
""").df()
```

```python
quant_messages_per_media_type_df
```
