## m) Identificar inconsistências entre os atributos (features)

### Inconsistências entre colunas `trava_zap`

```python
def has_trava_zap(texto):
  if not isinstance(texto, str):
    return False  # Garante que só analisa strings

  # Regra 1: muito longo
  if len(texto) > 10000:
    return True

  # Regra 2: caracteres invisíveis
  invisiveis = ''.join(chr(c) for c in range(0x200B, 0x200F + 1)) + '\uFEFF'
  if sum(texto.count(c) for c in invisiveis) > 100:
    return True

  # Regra 3: emojis repetidos
  emoji_pattern = re.compile("[\U0001F600-\U0001F64F\U0001F300-\U0001F5FF"
                              "\U0001F680-\U0001F6FF\U0001F1E0-\U0001F1FF]+", flags=re.UNICODE)
  emojis = emoji_pattern.findall(texto)
  if any(len(e) > 20 for e in emojis):
    return True

  # Regra 4: baixa diversidade
  unicos = set(texto)
  if len(unicos) < 10 and len(texto) > 500:
    return True

  return False
```

```python
df['has_travazap2'] = df['text_content_anonymous'].apply(has_trava_zap)
```

```python
from sklearn.metrics import accuracy_score

print("Acurácia:", accuracy_score(df['has_travazap2'], df['trava_zap']))
```

### Inconsistência na coluna `score_sentiment`

```python
from LeIA import SentimentIntensityAnalyzer

s = SentimentIntensityAnalyzer()

def recalculate_sentiment_score(text):
  if isinstance(text, str) and pd.notnull(text):
    score = s.polarity_scores(text)
    return score['compound']
  else:
    return 0
```

```python
df['new_sentiment_score'] = df['text_content_anonymous'].apply(recalculate_sentiment_score)
df['new_sentiment_score']
```
