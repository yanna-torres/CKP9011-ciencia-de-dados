## f) Identificar e listar as posições (células) contendo valores que não pertencem ao domínio (tipo de dados) esperado

Para verificar se há valores que não pertencem ao tipo de dado esperado por coluna, primeiro são removidos os valores nulos com `dropna()`, pois eles podem interferir na verificação dos tipos.

Em seguida, é feita uma varredura em cada coluna do dataframe limpo, coletando os tipos de dados presentes em cada uma.

```python
df_clean = df.dropna()
column_types = []
for column in df_clean.columns:
    existing_types = {type(value) for value in df_clean[column]}
    column_types.append((column, existing_types))
```

## g) Crie uma coluna chamada “caracteres” contendo a quantidade de caracteres da coluna “text”

É utilizada a função apply do Pandas em conjunto com uma função lambda, que aplica `len(x)` para medir o comprimento de cada string. Caso o valor da célula seja nulo, a função retorna `0`.

```python
df['caracteres'] = df['text_content_anonymous'].apply(lambda x: len(x) if pd.notnull(x) else 0)
df['caracteres']
```

```python
df['caracteres'].describe()
```

## h) Crie uma coluna chamada “words” contendo a quantidade de palavras da coluna “text”

Utilizamos a mesma lógica do item anterior, porém, ao invés de calcular o número de caracteres, calculamos o número de palavras. Para isso, a função `split()` é usada para separar o texto em palavras, e o método `len()` conta quantas palavras foram extraídas.

```python
df['words'] = df['text_content_anonymous'].apply(lambda x: len(x.split()) if pd.notnull(x) else 0)
df['words']
```

```python
df['words'].describe()
```

## i) Crie uma coluna chamada “viral” contendo o valor 0 se o texto da mensagem (valor do campo “text”) não for encontrado em outras linhas e 1, caso contrário

Para identificar se um texto é "viral", usamos a função `value_counts()` do Pandas para contar a ocorrência de cada valor na coluna `text_content_anonymous`.

Se o texto aparecer mais de uma vez no dataframe, o texto é considerado "viral" e recebe o valor `1`. Caso contrário, recebe o valor `0`.

```python
text_counts = df['text_content_anonymous'].value_counts()
df['viral'] = df['text_content_anonymous'].apply(lambda x: 1 if pd.notnull(x) and text_counts[x] > 1 else 0)
df['viral']
```

```python
print('Amount of viral texts:')
len(df[df['viral'] == 1])
```

## j) Crie uma coluna chamada “sharings” contendo a quantidade de vezes que o texto armazenado no atributo “text” aparece no dataset

Para contar o número de vezes que cada texto aparece no dataset, reutilizamos a variável `text_counts` do item anterior, que contém a contagem de ocorrências de cada valor na coluna `text_content_anonymous`

```python
df['sharings'] = df['text_content_anonymous'].apply(lambda x: text_counts[x] if pd.notnull(x) else 0) # utilizando a mesma variável do item anterior
df['sharings']
```

```python
print('Top 10 viral texts:')
text_counts.head(10)
```

## k) Crie uma coluna chamada “sentiment” contendo os valores: -1 para textos negativos, 0 para textos neutros e 1 para textos positivos

A coluna sentiment é criada com base nos valores da coluna `score_sentiment`, que contém os resultados de uma análise de sentimento.

O valor de sentiment é atribuído de acordo com o seguinte critério:

* 1 para textos com sentimento positivo (quando o valor de score_sentiment for maior ou igual que 0.05)

* 0 para textos com sentimento neutro (quando o valor de score_sentiment for entre -0.05 e 0.05)

* -1 para textos com sentimento negativo (quando o valor de score_sentiment for menor ou igual que -0.05)

```python
df['sentiment'] = df['score_sentiment'].apply(lambda x: 1 if x >= 0.05 else (-1 if x <= -0.05 else 0))
df['sentiment']
```

```python
df['score_sentiment'].describe()
```

## m) Identificar inconsistências entre os atributos (features)

Neste item, vamos verificar se existem inconsistências nas colunas do dataframe original. Será verificado as possíveis inconsistências seguintes:

* **Verificação da coluna `trava_zap`**: Verificar se os valores desta coluna estão corretos, garantindo que o valor True corresponde a textos que realmente contêm "trava-zaps" e o valor False para os demais.
* **Verificação de IDs duplicados**: Verificar se a coluna que armazena os IDs possui valores duplicados.
* **Verificação da coluna `has_media`**: Verificar se a coluna `has_media` está correta, ou seja, se um valor True indica a presença de mídia e False indica a ausência.
* **Verificação da coluna `has_media_url`**: Verificar se a coluna `has_media_url` está correta, ou seja, se True indica que uma URL de mídia está presente na coluna `media_url` e False indica sua ausência.

### Inconsistências na coluna `trava_zap`

```python
def has_trava_zap(message):
  if not isinstance(message, str):
    return False  # Garante que só analisa strings

  # Regra 1: muito longo
  if len(message) > 10000:
    return True

  # Regra 2: caracteres invisíveis
  invisible = ''.join(chr(c) for c in range(0x200B, 0x200F + 1)) + '\uFEFF'
  if sum(message.count(c) for c in invisible) > 100:
    return True

  # Regra 3: emojis repetidos
  emoji_pattern = re.compile("[\U0001F600-\U0001F64F\U0001F300-\U0001F5FF"
                              "\U0001F680-\U0001F6FF\U0001F1E0-\U0001F1FF]+", flags=re.UNICODE)
  emojis = emoji_pattern.findall(message)
  if any(len(e) > 20 for e in emojis):
    return True

  # Regra 4: baixa diversidade
  unique = set(message)
  if len(unique) < 10 and len(message) > 500:
    return True

  return False
```

```python
df['has_travazap2'] = df['text_content_anonymous'].apply(has_trava_zap)
```

```python
tp = ((df['has_travazap2'] == True) & (df['trava_zap'] == True)).sum()
tn = ((df['has_travazap2'] == False) & (df['trava_zap'] == False)).sum()
fp = ((df['has_travazap2'] == True) & (df['trava_zap'] == False)).sum()
fn = ((df['has_travazap2'] == False) & (df['trava_zap'] == True)).sum()

print(f"False Positives (FP): {fp}")
print(f"False Negatives (FN): {fn}")

accuracy = (tp + tn) / (tp + tn + fp + fn)
print(f"Accuracy: {accuracy:.2%}")
```

```python
mismatches = df[df['has_travazap2'] != df['trava_zap']]
print(f"Number of mismatches: {len(mismatches)}")
mismatches.head()
```

```python
df.drop(columns=['has_travazap2'], inplace=True)
```
