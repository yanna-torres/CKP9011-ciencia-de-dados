### f) Identificar e listar as posições (células) contendo valores que não pertencem ao domínio (tipo de dados) esperado

```python
expected_types = {
    'date_message': 'datetime',
    'id_member_anonymous': 'str',
    'id_group_anonymous': 'str',
    'media': 'str',
    'media_type': 'str',
    'media_url': 'str',
    'has_media': 'bool',
    'has_media_url': 'bool',
    'trava_zap': 'bool',
    'text_content_anonymous': 'str',
    'dataset_info_id': 'int',
    'date_system': 'datetime',
    'score_sentiment': 'float',
    'score_misinformation': 'float',
    'id_message': 'int',
    'message_type': 'str',
    'messenger': 'str',
    'media_name': 'str',
    'media_md5': 'str',
}
```

```python
from datetime import datetime

# Função auxiliar para detectar tipo datetime
def is_datetime(val):
    return isinstance(val, (pd.Timestamp, datetime))

# Armazenar valores inválidos
for col, expected_type in expected_types.items():
    invalid_mask = []

    for val in df[col]:
        if pd.isnull(val):
            invalid_mask.append(False)
        elif expected_type == 'datetime':
            invalid_mask.append(not is_datetime(val))
        elif isinstance(expected_type, tuple):  # ex: (int, float)
            invalid_mask.append(not isinstance(val, expected_type))
        else:
            invalid_mask.append(not isinstance(val, expected_type))

    invalid_mask = pd.Series(invalid_mask)

    if invalid_mask.any():
        print(f"\nColuna '{col}' contém {invalid_mask.sum()} valor(es) fora do tipo esperado:")
        print(df.loc[invalid_mask, col])
```
